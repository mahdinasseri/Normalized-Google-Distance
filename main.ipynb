```python
mean_Values = []
for n in range(10):
    print("Started:", n)
    element_list = get_random_words(n=4)
    matrix = pairwise_NGD(element_list, lan='en')
    m = pd.DataFrame(matrix)
    x = len(m)
    values = []
    for i in range(0,x):
        for j in range(i+1,x):
            #print(i,j, m.iloc[i,j])
            values.append(m.iloc[i,j])
    #print(values)
    mean_Values.append(np.mean(values))
```
